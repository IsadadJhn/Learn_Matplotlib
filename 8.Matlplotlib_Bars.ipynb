```python
import matplotlib.pyplot as plt
import numpy as np
```

**Vertikal/Default Bar**

```python
x = np.array(['A','B','C'])
y = np.array([10,15,20])
plt.bar(x,y)
plt.show()
```

**Horizontal Bar**

```python
x = np.array(['A','B','C'])
y = np.array([10,15,20])
plt.barh(x,y)
plt.show()
```

**Bar Color**

```python
# Color Names
x = np.array(['A','B','C'])
y = np.array([10,20,30])
plt.bar(x,y, color='hotpink')
plt.show()
```

```python
# Color Hex
x = np.array(['A','B','C'])
y = np.array([10,15,20])
plt.bar(x,y, color='#1fd655')
plt.show()
```

**Bar Width**

```python
x = np.array(['A','B','C'])
y = np.array([10,15,20])
plt.bar(x,y, color='#1fd655',width=0.3) #default value = 0.8
plt.show()
```

**Bar Height**

```python
x = np.array(['A','B','C'])
y = np.array([10,15,20])
plt.barh(x,y, color='#1fd655',height=0.2) #default value = 0.8
plt.show()
```
