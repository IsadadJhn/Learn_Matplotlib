**Matplotlib Subplot**

**with the subplot() function you can draw multiple plots in one figure**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Plots in one row (kiri-kanan)**

```python
#plot 1
x1 = np.array([0,1,2,3])
y1 = np.array([3,7,2,10])

plt.subplot(1,2,1) # --> 1 rows , 2 columns , and this plot is the first plot
plt.plot(x1,y1, marker='.',linestyle=':',color='g')
plt.grid()

#plot 2 
x2 = np.array([0,1,2,3])
y2 = np.array([80,60,20,10])
plt.subplot(1,2,2) # --> 1 rows, 2 columns, ,and this plot is the second plot
plt.plot(x2,y2,marker='.',linestyle=':',color='darkred')
plt.grid()
plt.show()
```

**Plots in 2 rows(atas-bawah)**

```python
#plot 1
x1 = np.array([0,1,2,3])
y1 = np.array([3,7,2,10])

plt.subplot(2,1,1) # --> 2 rows , 1 columns , and this plot is the first plot
plt.plot(x1,y1, marker='.',linestyle=':',color='g')
plt.grid()

#plot 2 
x2 = np.array([0,1,2,3])
y2 = np.array([80,60,20,10])
plt.subplot(2,1,2) # --> 2 rows, 1 columns, ,and this plot is the second plot
plt.plot(x2,y2,marker='.',linestyle=':',color='darkred')
plt.grid()
plt.show()
```

**Title**

```python
#Title for each plot
#plot 1
x1 = np.array([0,1,2,3])
y1 = np.array([3,7,2,10])

plt.subplot(1,2,1) # --> 1 rows , 2 columns , and this plot is the first plot
plt.plot(x1,y1, marker='.',linestyle=':',color='g')
plt.grid()
plt.title('Sales')
#plot 2 
x2 = np.array([0,1,2,3])
y2 = np.array([80,60,20,10])
plt.subplot(1,2,2) # --> 1 rows, 2 columns, ,and this plot is the second plot
plt.plot(x2,y2,marker='.',linestyle=':',color='darkred')
plt.grid()
plt.title('Income')
plt.show()
```

**Super Title**

```python
#Title to the entire figure
x1 = np.array([0,1,2,3])
y1 = np.array([3,7,2,10])

plt.subplot(1,2,1) # --> 1 rows , 2 columns , and this plot is the first plot
plt.plot(x1,y1, marker='.',linestyle=':',color='g')
plt.grid()
plt.title('Sales')
#plot 2 
x2 = np.array([0,1,2,3])
y2 = np.array([80,60,20,10])
plt.subplot(1,2,2) # --> 1 rows, 2 columns, ,and this plot is the second plot
plt.plot(x2,y2,marker='.',linestyle=':',color='darkred')
plt.grid()
plt.title('Income')
plt.suptitle('My Shop')
plt.show()
```
