```python
import matplotlib.pyplot as plt
import numpy as np
```

**Pie Charts**

```python
#By default the plotting of the first wedge starts from the x-axis and moves counterclockwise
#grafik dibaca dari sumbu x dan bergerak berlawanan arah jarum jam
x = np.array([45,35,25,15]) 
plt.pie(x)
plt.show()
```

**Labels**

```python
x = np.array([45,35,25,15]) 
mylabels = ['Banana','Kurma','Manggo','Jackfruit']
plt.pie(x,labels=mylabels)
plt.show()
```

**Start Angle**

```python
#default start angle is at x-axis(sumbu x = 0),but you can change with startangle parameters
x = np.array([45,35,25,15]) 
mylabels = ['Banana','Kurma','Manggo','Jackfruit']
plt.pie(x,labels=mylabels,startangle=180)
plt.show()
```

**Explode**

```python
#explode itu parameter buat ngedorong salah satu irisan pie chart keluar sedikit dari lingkaran utama.
x = np.array([45,35,25,15]) 
mylabels = ['Banana','Kurma','Manggo','Jackfruit']
myexplode = [0.3,0,0,0]
plt.pie(x,labels=mylabels,explode=myexplode)
plt.show()
```

**Shadow**

```python
#add a shadow to pie charts 
#menambahkan bayangan di diagram lingkaran (pie charts)
x = np.array([45,35,25,15]) 
mylabels = ['Banana','Kurma','Manggo','Jackfruit']
myexplode = [0.3,0,0,0]
plt.pie(x,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

**Color of each wedge (irisan)**

```python
x = np.array([45,35,25,15]) 
mylabels = ['Banana','Kurma','Manggo','Dragon Fruit']
mycolors = ['yellow','brown','green','red'] #make an array of colors
plt.pie(x,labels=mylabels,colors=mycolors,shadow=True) #set the colors of each wedge
plt.show()
```

**Legends**

```python
#to add a list of explanation for each wedge, use legend function
#untuk kasih keterangan pada setiap irisan ,gunakan fungsi legend
x = np.array([45,35,25,15]) 
mylabels = ['Banana','Kurma','Manggo','Dragon Fruit']
mycolors = ['yellow','brown','green','red'] #make an array of colors
plt.pie(x,labels=mylabels) #set the colors of each wedge
plt.legend()
plt.show()
```

**Legend with Header**

```python
x = np.array([45,35,25,15]) 
mylabels = ['Banana','Kurma','Manggo','Dragon Fruit']
mycolors = ['yellow','brown','green','red'] #make an array of colors
plt.pie(x,labels=mylabels) #set the colors of each wedge
plt.legend(title='type of fruits') #untuk nambahin title legend
plt.show()
```
