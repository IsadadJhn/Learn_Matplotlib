**Introduction Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Plotting (Menggambar data menjadi grafik)**

```python
xpoints = np.array([0,10])  # sumbu X (Horizontal) --> 0-10
ypoints = np.array([0,100]) # sumbu Y (Vertikal) --> 0-100

plt.plot(xpoints,ypoints) #membuat grafik
plt.show() #menampilkan grafik 
```

```python
xpoints = np.array([0,20])  # sumbu X (Horizontal) --> 0-6
ypoints = np.array([0,175]) # sumbu Y (Vertikal) --> 0-100

plt.plot(xpoints,ypoints)
plt.show() #menampilkan grafik 
```
