```python
import matplotlib.pyplot as plt
import numpy as np
```

**Scatter**

```python
#Scatter itu grafik yang ditampilkan dalam bentuk titik-titik/ grafik sebaran titik
#plot() --> untuk melihat ppola pertumbuhan
#scatter --> untuk melihat hubungan/korelasi
```

```python
#notes: banyak data pada variabel x & y harus sama
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) # --> banyak data = 13
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86]) # --> banyak data = 13
plt.xlabel("Age of Cars")
plt.ylabel("Speed of Cars")
plt.scatter(x,y)
plt.title('Correlation between age and speed of cars')
plt.show()
```

```python
# day one , the age and speed of 13 cars
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
# day two , the age and speed of 15 cars
x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x1,y1, color='blue', label='Day 1')
plt.scatter(x2,y2, color='red', label='Day 2')
plt.legend() # --> untuk menampilkan keterangan/label grafik
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y, c=colors)
plt.show()
```

**colormap/cmap**

```python
#colormap = gradasi warna yang dipakai untuk merepresentasikan nilai data
x = np.array([17,12,11,9,9,8,7,7,6,5,4,2,2]) # -->age of each car
y = np.array([86,77,78,85,87,87,86,88,86,99,94,111,103]) # --> speed of each car
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100]) # dot color
size = np.array([20,50,10,20,50,10,60,90,10,30,60,80,75]) # dot size
plt.scatter(x,y,c = colors,cmap='Spectral',s = size,alpha= 0.9) # c = color , cmap = colormap , s = dot size , alpha = transparency
plt.colorbar() # colormap dalam bentuk bar
plt.show()
```
